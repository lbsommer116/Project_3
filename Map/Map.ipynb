```python
import dash
from dash import dcc, html, Input, Output, ctx
import plotly.express as px
import pandas as pd

# Your Data
data = {
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Miami"],
    "State": ["NY", "CA", "IL", "TX", "AZ", "FL"],
    "Latitude": [40.7128, 34.0522, 41.8781, 29.7604, 33.4484, 25.7617],
    "Longitude": [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740, -80.1918],
    "Home Value 2018": [600000, 700000, 380000, 330000, 400000, 580000],
    "Home Value 2019": [620000, 720000, 390000, 340000, 410000, 590000],
    "Home Value 2020": [630000, 730000, 395000, 345000, 415000, 595000],
    "Home Value 2021": [640000, 740000, 398000, 348000, 420000, 598000],
    "Home Value 2022": [650000, 750000, 400000, 350000, 425000, 600000],
    "Home Value 2023": [660000, 760000, 405000, 355000, 430000, 605000],
    "Home Value 2024": [670000, 770000, 410000, 360000, 435000, 610000],
    "Rental Value 2018": [950, 1900, 2900, 3900, 4900, 5900],
    "Rental Value 2019": [975, 1950, 2950, 3950, 4950, 5950],
    "Rental Value 2020": [985, 1975, 2975, 3975, 4975, 5975],
    "Rental Value 2021": [995, 1990, 2990, 3990, 4990, 5990],
    "Rental Value 2022": [1000, 2000, 3000, 4000, 5000, 6000],
    "Rental Value 2023": [1010, 2010, 3010, 4010, 5010, 6010],
    "Rental Value 2024": [1020, 2020, 3020, 4020, 5020, 6020]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Initialize Dash App
app = dash.Dash(__name__)

# Dropdown options for dataset type
dataset_options = [
    {"label": "Home Value", "value": "Home Value"},
    {"label": "Rental Value", "value": "Rental Value"},
]

# Dropdown options for year
year_options = [
    {"label": "2018", "value": "2018"},
    {"label": "2019", "value": "2019"},
    {"label": "2020", "value": "2020"},
    {"label": "2021", "value": "2021"},
    {"label": "2022", "value": "2022"},
    {"label": "2023", "value": "2023"},
    {"label": "2024", "value": "2024"},
]

# Initial Figure (Default to Home Value 2018)
fig = px.scatter_mapbox(
    df,
    lat="Latitude",
    lon="Longitude",
    size="Home Value 2018",
    hover_name="City",
    hover_data={"State": True, "Home Value 2018": True, "Latitude": False, "Longitude": False},
    color="Home Value 2018",
    color_continuous_scale="Viridis",
    zoom=3,
    height=600,
)

# Set Mapbox Token
fig.update_layout(
    mapbox_style="carto-positron",
    mapbox_zoom=4,
    mapbox_center={"lat": 37.0902, "lon": -95.7129},
)

# Dash Layout
app.layout = html.Div(
    [
        html.H1("US Home & Rental Values Bubble Map"),
        dcc.Dropdown(
            id="dataset-dropdown",
            options=dataset_options,
            value="Home Value",
            placeholder="Select Dataset",
        ),
        dcc.Dropdown(
            id="year-dropdown",
            options=year_options,
            value="2018",
            placeholder="Select Year",
        ),
        dcc.Graph(id="map", figure=fig, clickData=None),
        dcc.Graph(id="line-graph", figure={}),
    ]
)

# Callback to update the map based on dropdown selection
@app.callback(
    Output("map", "figure"),
    [Input("dataset-dropdown", "value"), Input("year-dropdown", "value")],
)
def update_map(selected_dataset, selected_year):
    selected_column = f"{selected_dataset} {selected_year}"
    fig = px.scatter_mapbox(
        df,
        lat="Latitude",
        lon="Longitude",
        size=selected_column,
        hover_name="City",
        hover_data={"State": True, selected_column: True, "Latitude": False, "Longitude": False},
        color=selected_column,
        color_continuous_scale="Viridis",
        zoom=3,
        height=600,
    )
    fig.update_layout(
        mapbox_style="carto-positron",
        mapbox_zoom=4,
        mapbox_center={"lat": 37.0902, "lon": -95.7129},
    )
    return fig

# Callback to update the line graph based on map click
@app.callback(
    Output("line-graph", "figure"),
    [Input("map", "clickData"), Input("dataset-dropdown", "value")]
)
def update_line_graph(clickData, selected_dataset):
    if clickData:
        city = clickData["points"][0]["hovertext"]
        years = [str(year) for year in range(2018, 2025)]
        values = [df[f"{selected_dataset} {year}"][df["City"] == city].values[0] for year in years]

        fig = px.line(x=years, y=values, title=f"{selected_dataset} Over Time for {city}")
        fig.update_layout(xaxis_title="Year", yaxis_title=f"{selected_dataset} Value")
        return fig
    else:
        return {}

# Run App
if __name__ == "__main__":
    app.run(debug=True)
```
