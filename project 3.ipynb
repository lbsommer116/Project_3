```python
# Create function to take average of years in each dataframe
def calculate_yearly_averages(df):
    
    # Takes a DataFrame with columns containing date strings (YYYY-MM-DD format),
    # # calculates row-wise yearly averages, and returns a new DataFrame with only yearly averages.
    # # Parameters:
    #     df (pd.DataFrame): Input DataFrame with date columns.
    # Returns:
    #     pd.DataFrame: DataFrame with yearly averages only.
    # “”"
    df_copy = df.copy()
    # Extract unique years from column names
    years = sorted(set(col[:4] for col in df_copy.columns if col[:4].isdigit()))
    # Calculate row-wise yearly averages
    for year in years:
        regex_var = f'^{year}-'
        month_cols = df_copy.filter(regex=regex_var).columns
        if len(month_cols) > 0:
            df_copy[year] = df_copy[month_cols].mean(axis=1, skipna=True)
    # Drop monthly columns (YYYY-MM-DD format)
    df_avg = df_copy.drop(df_copy.filter(regex=r'^\d{4}-').columns, axis=1)
    return df_avg
```
