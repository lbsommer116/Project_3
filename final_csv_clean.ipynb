```python
import pandas as pd
import re as re
```

```python
df_newcon_count_avg = pd.read_csv("df_newcon_count_avg.csv")
    
# Round the data in those columns
year_columns = [col for col in df_newcon_count_avg.columns if re.fullmatch(r'\d{4}', str(col))]
df_newcon_count_avg[year_columns] = df_newcon_count_avg[year_columns].round().astype("Int64")

#Save and show dataframe
df_newcon_count_avg.to_csv("df_newcon_count_avg.csv", index=False)
df_newcon_count_avg.head()
```

```python
df_value_index_avg = pd.read_csv("df_value_index_avg.csv")
 
# Round the data in those columns
year_columns = [col for col in df_value_index_avg.columns if re.fullmatch(r'\d{4}', str(col))]
df_value_index_avg[year_columns] = df_value_index_avg[year_columns].round(2)

#Save and show dataframe
df_value_index_avg.to_csv("df_value_index_avg.csv", index=False)
df_value_index_avg.head()
```

```python
df_days_pending_avg = pd.read_csv("df_days_pending_avg.csv")

# Round the data in those columns
year_columns = [col for col in df_days_pending_avg.columns if re.fullmatch(r'\d{4}', str(col))]
df_days_pending_avg[year_columns] = df_days_pending_avg[year_columns].round(0).astype("Int64")

#Save and show dataframe
df_days_pending_avg.to_csv("df_days_pending_avg.csv", index=False)
df_days_pending_avg.head()
```

```python
df_market_index_avg = pd.read_csv("df_market_index_avg.csv")

# Round the data in those columns
year_columns = [col for col in df_market_index_avg.columns if re.fullmatch(r'\d{4}', str(col))]
df_market_index_avg[year_columns] = df_market_index_avg[year_columns].round(0).astype("Int64")

# Remove any rows with values less than 0
df_market_index_avg = df_market_index_avg[(df_market_index_avg[year_columns] >= 0).all(axis=1)]

#Save and show dataframe
df_market_index_avg.to_csv("df_market_index_avg.csv", index=False)
df_market_index_avg.head()
```
